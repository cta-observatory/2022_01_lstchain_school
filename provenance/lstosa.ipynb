# lstosa provenance

Let's explore the provenance files produced by the on-site pipeline `lstosa`


```python
log_file = "../data/provenance/DL2/20200117/v0.8.2/tailcut84_model1/log/calibration_to_dl2_01807_prov.log"
json_file = "../data/provenance/DL2/20200117/v0.8.2/tailcut84_model1/log/calibration_to_dl2_01807_prov.json"
```

The log file contains a Python dictionary per line, being the first line a complex structured Python dictionary

```python
with open(log_file) as f:
    content = f.read()
```

```python
print(content)
```

Let's take the first line, which contains the system configuration provenance of the session

```python
ll = content.split("\n")
```

```python
print(ll[0])
```

We can convert that first line intio a session Python dictionary

```python
import ast
session = ast.literal_eval(ll[0])
```

```python
session
```

```python
session.keys()
```

```python
print(session["observation_date"])
```

```python
session['system']['python']['version'] # add and/or remove dict keys to browse the dict 
```

```python
for package in session["system"]["python"]["packages"]:
    if package["name"] == "protozfits":
        print(package["name"], package["version"])
```

The JSON file is a smaller file which does not contain the system configuration provenance. It is mostly needed to produce full provenance graphs where all components are linked.

```python
import json
with open(json_file) as j:
    provdata = json.load(j)
```
